```python
r_cols = ['user_id','movie_id','rating']
ratings = pd.read_csv('C:\\Users\\mojiway\\Desktop\\tutorials\\data-science-frank-kane-udemy\\my-files-data-science-udemy-frank-kane\\DataScience-Python3\\ml-100k\\u.data', sep='\\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")
```

```python
m_cols = ['movie_id','title']
movies = pd.read_csv('C:\\Users\\mojiway\\Desktop\\tutorials\\data-science-frank-kane-udemy\\my-files-data-science-udemy-frank-kane\\DataScience-Python3\\ml-100k\\u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")
```

```python
ratings = pd.merge(movies,ratings)
```

```python
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'], columns=['title'],values='rating')
movieRatings.head()
```

### about the pivot above: 
Each row in the pivot can be coverted to a vector showing what movies each \
user has rated and what were the rates. which is useful for user based recommendation. 
also each column can be converted to a vector saying how many times each movie is rated\
and what were the rates, which is so useful for item-based recommendation.

```python
movieRatings.columns
```

```python
odysseyRatings = movieRatings['2001: A Space Odyssey (1968)']
odysseyRatings.head()
```

```python
similarMovies = movieRatings.corrwith(odysseyRatings)
```

### corrwith():
calculates the correlation of one column with all other columns and rates them

```python
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
```

```python
similarMovies.head()
```

```python
similarMovies.sort_values(ascending=False)
```

# part 040

the results of part 3 didnt come out so well. now we want to know why and how we can improve it.
in fact some of the movies where rated by very few people and when even one person rates a movie for 5 the average of the\
rate will be 5 and it get on the top of a movie that has hundreds of ratings by the average of 4.9 which is not what we want.
- here we need to get rid of the movies that were rated by less than 100 people.

```python
moviesStats = ratings.groupby('title').agg({'rating':[np.size, np.mean]})
moviesStats.head()
```

```python
popularMovies = moviesStats['rating']['size'] >=100
popularMovies.head()
```

```python
moviesStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = moviesStats[popularMovies].join(pd.DataFrame(similarMovies ,columns=['similarity']))
```
